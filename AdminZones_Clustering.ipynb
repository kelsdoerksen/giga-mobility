# Clustering based on Location
From MapBox: The data for each country is normalized within that country. Comparisons made over short timescales and short distances are more trustworthy than those made over long timescales and distances. Comparisons made over areas that have a high average activity are more trustworthy than those made over areas that have low average activity.

Investigating Clustering via administrative boundaries of AOI (hardcoded for Botswana as this is the data we have). Hard-coded to admin level 1, can update admin_lvl variable in Load Data section to specify administrative level of clustering

## Imports and Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

## Load Data

```python
# Loading BWA school geojson data, can replace in future with other AOIs we may have data for
sample_df = gpd.read_file('/Users/kelseydoerksen/Desktop/Giga/SchoolMapping/BWA/BWA_train.geojson')

# Specify admin level to cluster by
admin_lvl = 1

# Load country admin levels
country_adm_levels = gpd.read_file('/Users/kelseydoerksen/Desktop/Giga/SchoolMapping/BWA/geoBoundaries-BWA-ADM{}.geojson'.format(admin_lvl))
```

## Clustering via Country Admin Boundaries

```python
def cluster_via_adm(df, country_admin_boundaries, adm_level):
    """
    Function to cluster data via ADM boundaries
    :param: df: dataframe of school and non-school samples to cluster
    :param: country_admin_boundaris: geopandas dataframe of country boundary information
    :param: adm_level: administrative level to query for generating clusters
    
    :return: sample_df_with_clusters: df of samples with added administrative boundaries
    """
    # Transform sample df to EPSG4326
    sample_df_epsg4326 = df.to_crs('EPSG:4326')
    
    # Get intersection of sample df with admin boundaries
    sample_df_with_clusters = gpd.overlay(sample_df_epsg4326, country_admin_boundaries, how='intersection')
    
    return sample_df_with_clusters
```

```python
def plot_clusters(sample_df_with_clusters, adm_level):
    """
    Plot clustered data
    """
    # Plot clusters to visualize
    clusters = sample_df_with_clusters['shapeName'].unique()
    plt. figure(figsize=(8, 6))
    for i in range(len(clusters)):
        filt = sample_df_with_clusters[sample_df_with_clusters['shapeName'] == clusters[i]]
        plt.scatter(filt.geometry.x, filt.geometry.y, label=clusters[i])

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.title('Botswana ADM{} Boundary School/Non-School Sample Clustering'.format(adm_level))
    plt.show()
```

```python
clustered_samples = cluster_via_adm(sample_df, country_adm_levels, admin_lvl)
```

```python
plot_clusters(clustered_samples, admin_lvl)
```

```python
# Save clustered data to file
save_dir = ''
clustered_samples.to_csv('{}/BWA_train_adminlevel{}_clusters.csv'.format(save_dir, admin_lvl))
```
